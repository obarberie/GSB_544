# Task 1

## Task 1.1

First, the size of the bubbles is based on population. Second, the color of the bubbles is based on geographic location. Third, the Y of the visual represents Life Expectancy. Lastly, the x axis demonstrates income.

## Task 1.2

```python
import pandas as pd
import plotnine as p9
from plotnine import *
from matplotlib.ticker import FixedLocator
```

```python
q1df = pd.read_csv('q1data.csv')
```

## Task 1.3

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'tomato', 'europe':'gold'}

(ggplot(q1df, aes(x='income',
y='life_exp',
size='population',
fill='four_regions',
label='country'))+
geom_text(aes(x=1400, y = 30, label = "2010"), size = 200, color = "lightgrey", alpha = .1, ha='left', va="bottom") +
geom_point(color = "black", stroke = .2 )+
labs(x="Income", y="Life Expectancy")+
scale_x_continuous(limits = (500,64000),breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],trans='log2',labels=[500, 1000, 2000, 4000, 8000, '16k', '32k', '64k'])+
scale_y_continuous(limits=(20,90),breaks=range(20,91, 10))+
scale_size_continuous(range=(0, 20))+
theme(figure_size = (14,6)) +
scale_fill_manual(values=region_colors)
)
```

## Task 1.4

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'tomato', 'europe':'gold'}

(ggplot(q1df, aes(x='income',
y='life_exp',
size='population',
fill='four_regions',
label='country'))+
geom_text(aes(x=1400, y = 30, label = "2010"), size = 200, color = "lightgrey", alpha = .1, ha='left', va="bottom") +
geom_polygon( )+
labs(x="Income", y="Life Expectancy")+
scale_x_continuous(limits = (500,64000),breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],trans='log2',labels=[500, 1000, 2000, 4000, 8000, '16k', '32k', '64k'])+
scale_y_continuous(limits=(20,90),breaks=range(20,91, 10))+
scale_size_continuous(range=(0, 20))+
theme(figure_size = (14,6)) +
scale_fill_manual(values=region_colors)
)
```

Compared to the geom point visual created above, this geom_polygon is pretty ineffective in displaying anything of importance. Partially do to the parameters of the previous graph created with geom_point.

## Task 2.1

The y axis reflects imports as % of GDP. The x axis shows exports as a percent of GDP. The size of the dots shows energy. Lastly, the Color of the bubbles reflects what country they represent. 

## Task 2.2

```python
q2df = pd.read_csv('q2data.csv')
```

## Task 2.3

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'tomato', 'europe':'gold'}

q2df['four_regions'] = pd.Categorical(q2df['four_regions'])
q2df['exports'] = pd.to_numeric(q2df['exports'], errors='coerce')
q2df['imports'] = pd.to_numeric(q2df['imports'], errors='coerce')
q2df['energy'] = pd.to_numeric(q2df['energy'], errors='coerce')

(ggplot(q2df, aes(x='exports',
y='imports',
size='energy',
fill='four_regions',
label='country'))+
geom_text(aes(x=35, y = 20, label = "1997"), size = 250, color = "lightgrey", alpha = .1, ha='left', va="bottom") +
geom_point(color = "black", stroke = .4 )+
#labs(x="Income", y="Life Expectancy")+
scale_x_continuous(limits = (0,240),breaks=[20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220],labels=[20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220])+
scale_y_continuous(limits=(0,400),breaks=range(50, 400, 50))+
scale_size_continuous(range=(0, 20))+
theme(figure_size = (14,6)) +
#, panel_background = element_text(color = "White")+
scale_fill_manual(values=region_colors)
)
```

## Task 2.4

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'tomato', 'europe':'gold'}

q2df['four_regions'] = pd.Categorical(q2df['four_regions'])
q2df['exports'] = pd.to_numeric(q2df['exports'], errors='coerce')
q2df['imports'] = pd.to_numeric(q2df['imports'], errors='coerce')
q2df['energy'] = pd.to_numeric(q2df['energy'], errors='coerce')

(ggplot(q2df, aes(x='exports',
y='imports',
size='energy',
fill='four_regions',
label='country'))+
geom_text(aes(x=35, y = 20, label = "1997"), size = 250, color = "lightgrey", alpha = .1, ha='left', va="bottom") +
geom_violin()+
#labs(x="Income", y="Life Expectancy")+
scale_x_continuous(limits = (0,240),breaks=[20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220],labels=[20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220])+
scale_y_continuous(limits=(0,400),breaks=range(50, 400, 50))+
scale_size_continuous(range=(0, 20))+
theme(figure_size = (14,6)) +
scale_fill_manual(values=region_colors)
)
```

Comparing these two 1997 graphs, the second option is actually preferred. Due to the density of the contries being if similar export/import values, the violin chart does a better job at showing the differences between the 4 countries. With that said, the violin (second graph) does miss out on some key visualizations that is given by the geom point.

# Task 3

## Task 3.1

The y axis reflects GDP. The x axis shows internet users. The size of the dots shows income. Lastly, the Color of the bubbles reflects what country each bubble represents. 

## Task 3.2

```python
q3df = pd.read_csv('q3data.csv')
```

## Task 3.3

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'tomato', 'europe':'gold'}

(ggplot(q3df, aes(x='internet_users',
y='gdp',
size='income',
fill='four_regions',
label='country'))+
geom_text(aes(x=35, y = 20, label = "2001"), size = 250, color = "lightgrey", alpha = .1, ha='left', va="bottom") +
geom_point(color = "black", stroke = .4 )+
scale_x_continuous(limits = (0,90),breaks=[10, 20, 30, 40, 50, 60, 70, 80, 90],labels=[10, 20, 30, 40, 50, 60, 70, 80, 90])+
scale_y_continuous(limits=(0,150000),breaks=[200,500,1000,2000,5000,10000,20000,50000,100000,150000], trans='log', labels=[200,500,1000,2000,5000,"10k","20k","50k","100K", " "])+
scale_size_continuous(range=(0, 30))+
theme(figure_size = (25,12)) +
scale_fill_manual(values=region_colors)
)
```

## Task 3.4

```python
region_colors = {'africa':'deepskyblue', 'americas':'lime', 'asia':'tomato', 'europe':'gold'}

(ggplot(q3df, aes(x='internet_users',
y='gdp',
size='income',
fill='four_regions',
label='country'))+
geom_text(aes(x=35, y = 20, label = "2001"), size = 250, color = "lightgrey", alpha = .1, ha='left', va="bottom") +
geom_boxplot()+
scale_x_continuous(limits = (0,90),breaks=[10, 20, 30, 40, 50, 60, 70, 80, 90],labels=[10, 20, 30, 40, 50, 60, 70, 80, 90])+
scale_y_continuous(limits=(0,150000),breaks=[200,500,1000,2000,5000,10000,20000,50000,100000,150000], trans='log', labels=[200,500,1000,2000,5000,"10k","20k","50k","100K", " "])+
scale_size_continuous(range=(0, 30))+
theme(figure_size = (25,12)) +
scale_fill_manual(values=region_colors)
)
```

Compared to the geom_point, this geom_histogram has many pros and cons. As for pros, the first is the readability of the visualization. The seperation of country by internet users shows clear differences in the size of each box. The boxplot however, does not quite show the same relationship as the geom point because of how each country is in its own unique box.
